```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
x = pd.DataFrame(df.iloc[:,:-1])
y = pd.DataFrame(df.iloc[:,-1])
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.2,random_state=42)
lr = LinearRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
print(r2_score(y_test,y_pred))
```

```python
#fit yeo-johnson to make it a normal data
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer(method='yeo-johnson')
x_train_pt = pt.fit_transform(x_train)
x_test_pt = pt.transform(x_test)
lr.fit(x_train_pt,y_train)
y_pred_pt = lr.predict(x_test_pt)
print(r2_score(y_test,y_pred_pt))
u_df = pd.DataFrame({'cols':x_train.columns,'coef':pt.lambdas_})
X_train_pt1 = pd.DataFrame(x_train_pt,columns=x_train.columns)
```

Conditional number analysis(as there is no function for this so we have to made
a function)

```python
r = df.drop('median_value',axis=1)

def remove_features(df,threshold):
  cond_number = np.linalg.cond(df)
  if cond_number>threshold:
    return df

  no_features = df.shape[1]
  for i in range(no_features):
    temp_df = df.drop(df.columns[i],axis = 1 )
    temp_cond_number = np.linalg.cond(temp_df)
    if temp_cond_number < cond_number:
      return remove_features(temp_df,threshold)
  return df
no = 30
updated_df  = remove_features(r,no)
updated_df
```
